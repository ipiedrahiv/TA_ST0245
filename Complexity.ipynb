# 2 Examples

In this section, we will look at many different algorithms and plot graphs that compare the length of the imput to the time they take to run. This will let us visually verify their complexity. 

We will use the plot function in order to graph the different functions without having to repeat code.

```python
def plot(times, n, lab):
    plt.xlabel('') 
    plt.ylabel('Time Complexity') 
    plt.plot(range(1, n), times, label = lab) 
    plt.grid() 
    plt.legend() 
    plt.show()
```

## 2.1 Addition

```python
import time
import matplotlib.pyplot as plt
import random


def addition(summand_1, summand_2):
    return summand_1 + summand_2

times = []

for i in range(1, 500):
    start_time = time.time()
    addition(random.randint(0, 100), random.randint(0, 100))
    times.append(time.time() - start_time)
    
plot(times, 500, "Addition")
    
```

## 2.2 Binary Search

This algorithm is a way in which we can sort an indexable list by pivoting its items. Quick sort, both recurively, as shown below, or iteratively is a $O(log_2 n)$ which is the soncd fastest type of computations you will fins in this course. 

```python
import random
import time
import matplotlib.pyplot as plt


def binary_search(item_list,item):
    first = 0
    last = len(item_list)-1
    found = False
    while( first<=last and not found):
        mid = (first + last)//2
        if item_list[mid] == item :
            found = True
        else:
            if item < item_list[mid]:
                last = mid - 1
            else:
                first = mid + 1	
    return found

times = []

for i in range(1, 6000):
    array = [random.randint(0, x) for x in range(6000)]
    array.sort()
    start_time = time.time()
    binary_search(array, array[random.randint(0, i)])
    times.append(time.time() - start_time)
    
plot(times, 6000, "Binary Search")
```

## 2. Array Sum

```python
import time
import matplotlib.pyplot as plt
import random

def array_sum(arr):
    count = 0
    for i in range(len(arr)):
        count += arr[i]
        
times = []

for i in range(1, 500):
    array = [random.randint(0, x) for x in range(i)]
    start_time = time.time()
    array_sum(array)
    times.append(time.time() - start_time)
    
plot(times, 500, "Array sum")
```

## 2. Bubble Sort

Bubble sort is a sorting algorithm that compares the elements of an array by pairs, and is known to be a $O(n^2)$ complexity algorithm.

```python
import time
import matplotlib.pyplot as plt
import random

def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(0, len(arr)-i-1):
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]

times = []

for i in range(1, 500):
    array = [random.randint(0, x) for x in range(i)]
    start_time = time.time()
    bubble_sort(array)
    times.append(time.time() - start_time)
    
plot(times, 500, "Bubble sort")
```
