## Ambientes Virtuales

- Son espacios aislados en los que puedes descargar todos los modulos de Python que quieras sin tener miedo.

    * Python tiene su propia forma de manejar y resolver paquetes, pero a veces toma decisiones raras sobre donde guardarlos por defecto.
    * Tu sistema operativo usa Python, y algunos modulos, no quieres dañar lo que tu sistema operativo necesita.
    
    
- Usamos Conda para crear y manejar estos ambientes virtuales, para descargar conda

    * Si usas [MacOS](https://docs.anaconda.com/anaconda/install/mac-os/)
    * Si usas [Linux](https://docs.conda.io/projects/conda/en/latest/user-guide/install/linux.html)
    * Si usas [Windows](https://docs.conda.io/projects/conda/en/latest/user-guide/install/windows.html)
 
 
- Ahora vamos a crear nuestro primer ambiente virtual, le vamos a poner de nombre isaeslamejormonitora y va a correr en python 3.6, para lograr eso corremos el siguiente comando desde la terminal

<blockquote><font color="gray" style="font-family:helvetica" size=3>conda create --name isaeslamejormonitora python=3.6</font></blockquote>

<blockquote><font color="gray" style="font-family:helvetica" size=3>conda create -n isaeslamejormonitora python=3.6</font></blockquote>

- Vamos a entrar a nuestro ambiente corriendo el siguiente comando

<blockquote><font color="gray" style="font-family:helvetica" size=3>conda activate isaeslamejormonitora</font></blockquote>

- Ahora vamos a instalar el manejador de paquetes de python, pip, corriendo el siguiente comando

<blockquote><font color="gray" style="font-family:helvetica" size=3>conda install pip</font></blockquote>

## ¿Como Puedo Resolver el Taller de la Semana 1 en Python?

- La respuesta corta es usando secuencias, condicionales, ciclos y objetos.

```python
class SammyElHeladero():
    
    # Constructor
    def __init__(self, amigos, buen_humor):
        self.amigos = amigos
        self.buen_humor = buen_humor
        
    # Funciones de clase
    def hacer_helado(self):
        print("Echarle agua")
        print("Echarle risa")
        
        if(buen_humor):
            print("Echarle leche")
            print("Echarle nueces molidas")
            print("Echarle un poco de brisa")
            
    def darle_helado_a_los_animales_de_africa(self):
        for animal in amigos:
            print(f"Sammy le dio helado al {animal}")
```

```python
amigos = ['leon', 'elefante', 'mono']
buen_humor = True

sammy = SammyElHeladero(amigos, buen_humor)

sammy.hacer_helado()
print('\n')
sammy.darle_helado_a_los_animales_de_africa()
```
